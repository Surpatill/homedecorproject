```python
data_path=r'D:\homedecor\archive\Super_Store_data.csv'
data=pd.read_csv(data_path,encoding='utf=8',encoding_errors='ignore')
```

```python
category_sales=data.groupby('Category')['Sales'].sum().reset_index()
category_sales=category_sales.sort_values('Sales',ascending=False)
print(category_sales.head(5))
```

```python
product_sales = data.groupby('Product Name')['Sales'].sum().reset_index()
product_sales = product_sales.sort_values('Sales', ascending=False)
print(product_sales)
```

```python
plt.bar(category_sales['Category'], category_sales['Sales'])
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Top-Selling Categories')
plt.show()
```

```python
plt.bar(Sub_Category_sales['Sub-Category'], Sub_Category_sales['Sales'])
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.title('Top-Selling Sub-Categories')
plt.show()
```

```python
top_products = product_sales.nlargest(30, 'Sales') 
plt.figure(figsize=(15,8))  
plt.bar(top_products['Product Name'], top_products['Sales'])
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.title('Top-Selling Products')
plt.xticks(rotation=90, fontsize=8) 
plt.gca().tick_params(axis='x', labelsize=8)  
plt.tight_layout()  
plt.show()
```

```python
customer_segments = data['Segment'].value_counts()
print(customer_segments)
plt.bar(customer_segments.index, customer_segments.values)
plt.xlabel('Segment')
plt.ylabel('Count')
plt.title('Customer Segmentation')
plt.show()
```

```python
top_30_sales_trends = data.nlargest(30, 'Sales')
plt.figure(figsize=(12,6))
plt.plot(top_30_sales_trends['Order Date'], top_30_sales_trends['Sales'], marker='o')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Top 30 Sales Trends by Order Date')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
